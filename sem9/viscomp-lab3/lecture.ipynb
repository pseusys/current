## Histograms

```python
from typing import List

import matplotlib.pyplot as plt
import numpy as np


def draw_histogram(image: List[List[int]]):
    maximum = max([max(l) for l in image]) + 1
    hist = [0] * maximum
    for i in image:
        for j in i:
            hist[j] += 1
    plt.bar(list(range(maximum)), hist)
    plt.show()
```

```python
draw_histogram([[0] * 10] + [[255] * 10] * 3 + [[0] * 10])
```

```python
draw_histogram([[0] * 5] + [[255] * 5] * 3 + [[0] * 5] * 2)
```

```python
image = [list()] * 26
for i in range(len(image)):
    image[i] += [i * 10 + i] * 10
draw_histogram(image)
```
